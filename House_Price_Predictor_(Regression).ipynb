**Step 1: The "Why" Before the "How" (Theory)**

Before we code, let's talk about Exploratory Data Analysis (EDA). In BDA, your model is only as good as the data you feed it. If you feed it "garbage," you get "garbage" out.

1. *Data Cleaning*
We check for Null Values (missing data). If a house has "0" bedrooms in the data but is 5000 sq. ft., thatâ€™s a mistake. We either fix it or drop it

2. *Correlation*
We want to see which features actually affect the price. Does the "Zipcode" matter more than "Square Footage"? We use a Correlation Matrix to find a number between -1 and 1.

*   1: Perfectly positive (as Square Feet goes up, Price goes up).
*   0: No relationship.
*  -1: Perfectly negative (as Age of house goes up, Price might go down).


**Step 2: Letâ€™s Code (Day 1: Loading & EDA)**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
url = "https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv"
df = pd.read_csv(url)

# 2. Look at the first 5 rows
print("--- First 5 Rows ---")
display(df.head())

# 3. Check for missing values (The 'Cleaning' part)
print("\n--- Missing Values ---")
print(df.isnull().sum())

# 4. Statistical Summary
print("\n--- Summary Stats ---")
display(df.describe())

# 5. Visualizing Correlation (The 'Detective' part)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()

# 2. Visualizing Correlation
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.1f')
plt.title("Correlation Heatmap (Numeric Only)")
plt.show()

# 3. Let's see the strongest relationship: Price vs Sqft_Living
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df)
plt.title('Relationship: Sqft Living vs Price')
plt.xlabel('Square Feet')
plt.ylabel('Price (in millions)')
plt.show()
```
